```python
from kpi.main import main
import pandas as pd
import numpy as np
```

```python
import random

# Creating a date range
date_range = pd.date_range(start='1/1/2022', end='12/31/2023')

# Creating random SKUs
skus = ['SKU' + str(i) for i in range(1, 11)]

# Creating random store names
stores = ['Store' + str(i) for i in range(1, 11)]

# Creating random sales data
sales = [random.randint(0, 100) for _ in range(len(date_range)*len(skus)*len(stores))]

# Creating random stock data
stock = [random.randint(0, 200) for _ in range(len(date_range)*len(skus)*len(stores))]

# Creating random price data
prices = [random.uniform(10, 50) for _ in range(len(date_range)*len(skus)*len(stores))]

# Creating a data frame
data = pd.DataFrame({
    'date': date_range.tolist() * len(skus) * len(stores),
    'sku': skus * len(date_range) * len(stores),
    'store': stores * len(date_range) * len(skus),
    'sales': sales,
    'stock': stock,
    'price': prices
})

# Splitting the data into our_data and enemy_data
our_data = data.sample(frac=0.5)
enemy_data = data.drop(our_data.index)

# Creating a price dictionary
price_dict = {sku: random.uniform(10, 50) for sku in skus}

# Creating a quantity dictionary
quantity_dict = {sku: random.randint(50, 200) for sku in skus}
```

```python
our_data_dict = {'sales': our_data, 'stock': our_data}
```

```python
kpi_dict = main({'sales': our_data, 'stock': our_data}, {'sales': enemy_data, 'stock': enemy_data}, data, price_dict, quantity_dict)
```
